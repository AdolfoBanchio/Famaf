```python
import numpy as np
import random as rd
```

```python
def ejercicio_2(n_sims):
    A_wins = 0
    for _ in range(n_sims):
        u1 = rd.random()
        u2 = rd.random()
        if u1 > 0.6 or u2 > 0.6:
            A_wins += 1
    
    return A_wins/n_sims

n = 1000000
print(ejercicio_2(n))
```

```python
def integral_monte_carlo (funciong, a, b , Nsim):
    integral = 0
    for i in range(Nsim):
        integral += funciong(a + (b-a) * rd.random())
    return integral * (b-a) / Nsim

def integral_monte_carlo_0_inf(funciong, Nsim):
    integral = 0
    for i in range(Nsim):
        u1 = rd.random()
        integral += 1/u1**2 * funciong(1/u1 - 1)
    return integral / Nsim

def integral_mult_monte_carlo(funciong, a, b, c, d, Nsim):
    integral = 0
    for i in range(Nsim):
        integral += funciong(a + (b-a) * rd.random(), c + (d-c) * rd.random())
        
    return integral / Nsim

def funcion_a(x,y):
    return 1 - np.exp(-(x+y))

def funcion_b(x):
    return x**2 * np.exp(-x**2)

for i in range(3,7):
    n = 10**i
    print(f"===== n = {n} =====")
    print(f"aproximacion a) = {integral_mult_monte_carlo(funcion_a, 0, 1, 0, 1, n)}")
    print(f"aproximacion b) = {integral_monte_carlo_0_inf(funcion_b, n) - integral_monte_carlo(funcion_b, 0, 1, n)}")
```
