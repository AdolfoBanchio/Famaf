```python
import numpy as np
import random as rd
```

```python
def ejercicio_2(nsims):
    wins_A = 0

    for i in range(nsims):
        # jugada 1
        uA = rd.random()
        vB = rd.random()
        if uA > 0.5 and vB < 0.5:
            wins_A += 1
        elif (uA < 0.5 and vB < 0.5) or (uA > 0.5 and vB > 0.5): 
            # empate 1 jugada, se juega otra
            uA = rd.random()
            vB = rd.random()
            if uA > 0.5 and vB < 0.5:
                wins_A += 1
    
    return wins_A/nsims

for i in range (2,7):
    n = 10**i
    print(f"===== n = {n} =====")
    print(f"Probabilidad de que A gane en primera o segunda jugada: {ejercicio_2(n)}")
```

```python
def integral_monte_carlo (funciong, a, b , Nsim):
    integral = 0
    for i in range(Nsim):
        integral += funciong(a + (b-a) * rd.random())
    return integral * (b-a) / Nsim

def integral_monte_carlo_0_inf(funciong, Nsim):
    integral = 0
    for i in range(Nsim):
        u1 = rd.random()
        integral += 1/u1**2 * funciong(1/u1 - 1)
    return integral / Nsim

def funcion_a(x):
    return x/(x- np.exp(x))

def funcion_b_alt(x):
    return x**3 * np.exp(-x**3)

for i in range (2,7):
    n = 10**i
    print(f"===== n = {n} =====")
    print(f"Aproximacion a) : {integral_monte_carlo(funcion_a, -3, 3, n)}")
    print(f"alternativa b) : {integral_monte_carlo(funcion_b_alt,-1,0,n)+integral_monte_carlo_0_inf(funcion_b_alt, n)} ")
```
