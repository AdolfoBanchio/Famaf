```python
import random as rd
import numpy as np
```

```python
def integral_monte_carlo_0_inf(funciong, Nsim):
    integral = 0
    for i in range(Nsim):
        u1 = rd.random()
        integral += 1/u1**2 * funciong(1/u1 - 1)
    return integral / Nsim
```

```python
def funcion_ej3(x):
    return 1/(((x+1)**2) * np.log(x+2))

def ejercicio_3():
    for i in range(3,6):
        n = 10**i
        resultado = integral_monte_carlo_0_inf(funcion_ej3, n)
        print(f"===== N = {n} =====")
        print(f"Aproximacion: {resultado}")

ejercicio_3()
```
