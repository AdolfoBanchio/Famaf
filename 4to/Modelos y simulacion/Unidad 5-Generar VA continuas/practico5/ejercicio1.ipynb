```python
import numpy as np
import random as rd
import matplotlib.pyplot as plt
```

Debo buscar la funcion de distribucion acomulada F y proceder usando el metodo de transformada inversa. 

Para este caso veo que cuando x = 2 debe valer 0 y cuando x = 6 debe valer 1. 

$$
\text{si}\quad  2 \leq  x \leq 3 , \\
F(X) = \int_{-\inf}^{2} 0 dt + \int_{2}^{x} \frac{t-2}{2} = \frac{x^2}{4} - x + 1\\
$$
$$
\text{si}\quad  3 \leq  x \leq 6 , \\
F(X) = \int_{-\inf}^{2} 0 dt + \int_{2}^{3} \frac{t-2}{2} dt + \int_{3}^{x} \frac{2 - \frac{t}{3}}{2} = \\
= \frac{-x^2}{12} + x - 2
$$


Luego la funcion de distribucion acumulada es:
$$
F(x) = \begin{cases} 
        0 \text{ si } x < 2 \\
        \frac{x^2}{4} - x + 1 \quad \text{ si } 2 \leq x \leq 3 \\ 
        \frac{-x^2}{12} + x - 2 \quad \text{ si } 3 \leq x \leq 6  \\
        1 \text{ si } x \gt 6
        \end{cases}

$$

Como F(3) = 1/4 , entonces en elalgoritmo que generara a X elegira segun el valor de u generado. 

si u < 1/4 entonces x = $F^{-1}(u) = 2 + \sqrt{4u}$

sino x = $6 -2 \sqrt{3(1-x)}$

```python
def tinversa():
    u = rd.random()
    if u < 1/4:
        x = 2 + np.sqrt(4*u)
    else:
        x = 6 - 2*np.sqrt(3*(1-u))
    return x
```

```python
# generate a plot for the inverse transformation method
x = [tinversa() for i in range(10000)]

plt.hist(x, bins=100, density=True)
plt.show()
```
