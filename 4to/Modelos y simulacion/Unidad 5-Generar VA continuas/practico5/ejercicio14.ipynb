```python
import numpy as np
import random as rd
```

```python
def eventos_poisson(lamda, T):
    t = 0
    NT = 0
    Eventos = []
    while t < T:
        U = 1 - rd.random()
        t += - np.log(U) / lamda
        if t <= T:
            NT += 1
            Eventos.append(t)
    return NT, Eventos
```

En este caso quiero saber cuantos colectivos llegaron en 1 hora. Y de esos colectivos simular la capacidad que tenian y de ahi sabre la cantidad de aficionados que llegaron en t horas.

```python
def hinchas_t_horas(lamda, T):
    NT, Eventos = eventos_poisson(lamda, T)

    total_hinchas = 0
    for i in range(NT):
        total_hinchas += rd.randint(20,41) # excluye el 41
    return total_hinchas
```

```python
lamda = 5
T = 1

total_hinchas = hinchas_t_horas(lamda, T)

print("Total de hinchas en 1 hora: ", total_hinchas)
```
