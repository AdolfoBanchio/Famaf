```python
def F_x(x, lamda):
    atg = np.arctan(x/lamda)
    return atg *(1/np.pi) + 0.5

def F_inv_x(x, lamda):
    tan = math.tan((x - 0.5) * math.pi)
    return lamda * tan
```

```python
def tinversa_x(lamda):
    u = rd.random()
    return F_inv_x(u, lamda)
```

```python
# calcular la probabilidad de que caiga en el intervalo (-lamda, lamda)

lamdas = [1, 2.5, 0.3]

exitos = [0, 0, 0]

nsims = 10000

for i in range(len(lamdas)):
    for j in range(nsims):
        if -lamdas[i] < tinversa_x(lamdas[i]) < lamdas[i]:
            exitos[i] += 1

print(f"probabilidad lambda = 1: {exitos[0]/nsims}")
print(f"probabilidad lambda = 2.5: {exitos[1]/nsims}")
print(f"probabilidad lambda = 0.3: {exitos[2]/nsims}")
```
