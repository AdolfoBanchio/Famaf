```python
import numpy as np
import random as rd
```

[explicacion](../generar_VA_continuas.md#generación-de-un-proceso-de-poisson)

```python
def eventos_poisson(lamda, T):
    t = 0
    NT = 0
    Eventos = []
    while t < T:
        U = 1 - rd.random()
        t += - np.log(U) / lamda
        if t <= T:
            NT += 1
            Eventos.append(t)
    return NT, Eventos
```

```python
lamda = 2
T = 3

NT, Eventos = eventos_poisson(lamda, T)

print("Numero de eventos: ", NT)
for i in range(NT):
    print("Evento ", i+1, ":", Eventos[i])
```
