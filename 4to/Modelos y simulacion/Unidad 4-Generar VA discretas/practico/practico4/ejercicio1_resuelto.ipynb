```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generar_baraja():
    mazo = np.arange(1, 101)
    np.random.shuffle(mazo)
    return mazo

def calcular_exitos_baraja():
    # generar un mazo con cartas del 1 al 100 y barajarlo
    mazo = generar_baraja()
    exitos = 0
    for i in range(len(mazo)):
        if mazo[i] == i + 1:
            exitos += 1

    return exitos
```

```python

def calcular_esperanza(nsims):
    # un exito es que la i-esima carta sea el numero i
    """ suma = 0
    for i in range(nsims):
        suma += calcular_exitos_baraja()
    
    return suma /nsims
    """
    sumas = [calcular_exitos_baraja() for i in range(nsims)]
    return np.mean(sumas), np.var(sumas)

def calcular_varianza(nsims):
    # valor teorico = 1, mu = 1.
    suma = 0
    for i in range(nsims):
        suma += (calcular_exitos_baraja() - 1)**2
    return suma / nsims
```

```python
# calculo de probabilidades 
def ejercicio1_a_i(nsims, r):
    # probabilidad de que las r primeras sean coincidencias
    exitos = 0
    for i in range(nsims):
        mazo = generar_baraja()
        if np.all(mazo[:r] == np.arange(1, r+1)):
            exitos += 1

    return exitos/nsims

def ejercicio1_a_ii(nsims, r):
    # debe haber exactamente r coincidencias y deben ser las primeras r 
    exitos = 0
    for i in range(nsims):
        mazo = generar_baraja()
        if np.all(mazo[:r] == np.arange(1, r+1)) and np.sum(mazo == np.arange(1, 101)) == r:
            exitos += 1
           
    return exitos/nsims 
```

```python
print(f"===== Valores Teoricos =====")
print(f"Esperanza: 1")
print(f"Varianza: 1")
print("----- Probabilidades -----")
print(f"i: (n-r)!/n! = (100-10)!/100! = 90!/(100!) = 1,59195994 x 10^-20 ~ 0")
print(f"ii: 1/r! sum(k=0 a n-r) -1^k/k! --> 1/10! sum(k=0 a 90) -1^k/k! = 1/10! (e^-1) = 0.")
for i in range(2,6):
    n=10**i
    print(f"===== n = {n} =====")
    esp, var = calcular_esperanza(n)
    print(f"Estimacion esperanza: {esp}")
    print(f"Estimacion varianza: {var}")
    print("----- Estimacion probabilidades -----")
    print(f"Estimacion i: {ejercicio1_a_i(n, 10)}")
    print(f"Estimacion ii: {ejercicio1_a_ii(n, 10)}")
```
