```python
import numpy as np
import random as rd
import time
```

Por el sig resultado teorico



Luego si tengo que 
$$
S = \sum_{i=1}^{N} f(i) \\ \text{Y quiero estimar S, entonces} \\
S = N \cdot \frac{1}{N} \sum_{i=1}^{N} f(i) = N \cdot E[g(X)] 
$$


```python
N = 10000

def g(x):
    return np.exp(x/N)
```

```python
"""
Primero voy a realizar el calculo de la estimacion de la Esperanza
y luego a ese resultado al multiplicarlo por N me dara el resultado esperado.
"""

def estimacion_s(nsims):
    suma = 0
    for i in range(nsims):
        x = rd.randint(1,N)
        suma += g(x)
    e_esp = suma/nsims
    return e_esp*N

def k_terminos(k):
    suma = 0
    for i in range(1,k+1):
        suma += g(i)
    return suma
```

```python
nsims = 100

# contar el tiempo de ejecucion de cada funcion
print(f"Estimacion de S con {nsims} numeros aleatorios:")
start_time1 = time.time()
est_s = estimacion_s(nsims)
stop_time1 = time.time()
print(est_s, "Tiempo de ejecucion: ", stop_time1-start_time1)

print(f"Caluclo de los primeros {nsims} terminos:")
start_time2 = time.time()
k_t = k_terminos(nsims)
stop_time2 = time.time()
print(k_t, "Tiempo de ejecucion: ", stop_time2-start_time2)

print(f"Valor real de S: {k_terminos(N)}")

```
