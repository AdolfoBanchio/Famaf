```python
import numpy as np
import random as rd
```

```python
def generate_data():
    n = 1
    mean = np.random.normal(0, 1)
    Scuad = 0 # S^2(1) = varianza muestral

    while n <= 100 or np.sqrt(Scuad / n) > 0.01:
        n += 1
        x = np.random.normal(0, 1)
        old_mean = mean
        mean = mean + (x - mean) / n
        Scuad = (1 - 1/(n-1)) * Scuad + n*(mean - old_mean)**2

    return n, mean, Scuad
```

```python
n, mean, Scuad = generate_data()
print(f"Numero de datos generados = {n}")
print(f"Media muestral = {mean}")
print(f"Varianza muestral = {Scuad}")
```
