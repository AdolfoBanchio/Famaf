```python
import random as rd
import numpy as np
import time
```

```python
def rejection():
    c=0.14
    p=[0.11,0.14,0.09,0.08,0.12,0.1,0.09,0.07,0.11,0.09]
    while True:
        y=int(10*rd.random())+1 #Y~U[1,10], q_i=P(Y=y_j) = 1/10 = 0.1 = qmin
        u=rd.random()
        if u<(p[y-1]/c):
            return y

#Notar que en realidad c=pmax/qmin = 0.14/0.1= 0.014. Pero como dsps en
#la condicion del if lo multiplico por qy=0.1 me estaría quedando siempre 0.14

```

```python
def transf_inv():
    u=rd.random()
    if u<0.14:
        return 2
    elif u<0.26:
        return 5
    elif u<0.37:
        return 1
    elif u<0.48:
        return 9
    elif u<0.58:
        return 6
    elif u<0.67:
        return 10
    elif u<0.76:
        return 7
    elif u<0.85:
        return 3
    elif u<0.93:
        return 4
    else:
        return 8
```

```python
def urna():
    #Creo el arreglo
    a=[]
    p=[0.11,0.14,0.09,0.08,0.12,0.1,0.09,0.07,0.11,0.09]
    for j in range(len(p)):
        for k in range(int(p[j]*100)):
            a.append(j+1)
    u=rd.random()
    return a[int(u*100)]
```

```python
urna()
```

```python
#Simulacion que permite ver que la v.a generada es la esperada.
def simulacion_1(n):
    start_time = time.time()
    a=np.zeros(10) #Cada elemento corresponde a las ocurrencias de su posición.
    for j in range(n):
        u=rejection()
        a[u-1]+=1 
    a=a/n
    end_time = time.time()
    return (a,end_time-start_time)
```

```python
#Simulacion que permite ver que la v.a generada es la esperada.
def simulacion_2(n):
    start_time = time.time()
    a=np.zeros(10) #Cada elemento corresponde a las ocurrencias de su posición.
    for j in range(n):
        u=transf_inv()
        a[u-1]+=1 
    a=a/n
    end_time = time.time()
    return (a,end_time-start_time)
    
```

```python
def simulacion_3(n):
    start_time = time.time()
    a=np.zeros(10) 
    for j in range(n):
        u=urna()
        a[u-1]+=1 
    a=a/n
    end_time = time.time()
    return (a,end_time-start_time)
```
